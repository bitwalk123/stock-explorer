# ダミーの空のデータフレームを用意

```python
df = pd.DataFrame()
df
```

# 時系列データフレームの生成

```python
for r in range(len(df0)):
    series = df0.iloc[r]
    ts = pd.to_datetime(series['Time'])
    price = series['Price']

    if len(df) == 0:
        # 最初のみデータフレームを（再）生成
        df = pd.DataFrame({'Price': [price]}, index=[ts])
    else:
        # データの追加
        df.loc[ts] = price
```
