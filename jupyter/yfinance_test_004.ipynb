```python
import datetime as dt
import yfinance as yf
yf.set_tz_cache_location('Asia/Tokyo')
```

```python
ticker = yf.Ticker('USDJPY=X')
```

```python
delta = dt.timedelta(days=1)
end = dt.date.today() + delta
start = end - 2 * delta

ticker.history(start=start, end=end, interval='1m')
```
