```python
dt_now = dt.datetime.now(ZoneInfo("Asia/Tokyo"))
dt_now
```

```python
y = dt_now.year
m = dt_now.month
d = dt_now.day
dt_start_1 = pd.to_datetime('%4d-%02d-%02d 09:00:00+09:00' % (y, m, d))
dt_end_1 = pd.to_datetime('%4d-%02d-%02d 11:30:00+09:00' % (y, m, d))

dt_start_2 = pd.to_datetime('%4d-%02d-%02d 12:30:00+09:00' % (y, m, d))
dt_end_2 = pd.to_datetime('%4d-%02d-%02d 15:00:00+09:00' % (y, m, d))
```

```python
df[df.index <= dt_end_1]
```

```python
df[df.index >= dt_start_2]
```

```python
df[(df.index <= dt_end_1) | (df.index >= dt_start_2)]
```

```python
dt_list_0 = pd.date_range(dt_start_1, periods=360, freq='min')
len(dt_list_0)
```

```python
dt_list = dt_list_0[(dt_list_0 <= dt_end_1) | (dt_list_0 >= dt_start_2)]
n = len(dt_list)
dt_list
```
