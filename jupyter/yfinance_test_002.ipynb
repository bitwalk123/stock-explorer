```python
import datetime as dt
import mplfinance as mpf
import yfinance as yf

#start = dt.date(2024, 7, 29)
end = dt.date.today()
delta = dt.timedelta(days=1)
start = end - delta

ticker = 'USDJPY=X'
df = yf.download(ticker, start, end, interval='1m')

#mpf.plot(df, type='candle', style='binance', figratio=(12, 4))
mpf.plot(df.tail(6*60), type='candle', style='binance', figratio=(12, 4))
df
```
