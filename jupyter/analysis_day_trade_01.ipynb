```python
pkl_1 = '8035.T_2024-03-25_2024-03-26_1m.pkl'
pkl_2 = '8035.T_2024-03-26_2024-03-27_1m.pkl'
pkl_3 = '8035.T_2024-03-27_2024-03-28_1m.pkl'
pkl_4 = '8035.T_2024-03-28_2024-03-29_1m.pkl'
pkl_5 = '8035.T_2024-03-29_2024-03-30_1m.pkl'
list_pkl = [pkl_1, pkl_2, pkl_3, pkl_4, pkl_5]
#list_pkl = [pkl_2, pkl_3, pkl_4]

title = '8035_2024_03-25--03-29'
xrange=['2024-03-25 08:00:00+09:00', '2024-03-29 16:00:00+09:00']
```

```python
df_result = pd.DataFrame()
for pkl in list_pkl:
    file = os.path.join(dir_path, pkl)
    df = pd.read_pickle(file)
    if len(df_result) == 0:
        df_result = df
    else:
        df_result = pd.concat([df_result, df])
df_result
```

```python
import matplotlib.pyplot as plt
import mplfinance as mpf

plt.rcParams["figure.figsize"] = (15, 5)
fig, ax = plt.subplots()
ax.set_title(title)

mpf.plot(
    df_result,
    xlim=xrange,
    type='line',
    style='binance',
    ax=ax
)
plt.grid(linestyle=':')
plt.subplots_adjust(left=0.06, right=0.995, bottom=0.2, top=0.93)
plt.savefig('output/%s.png' % title)
plt.show()
```
