```python
from functions.resources import get_connection
from PySide6.QtSql import QSqlQuery

code = 1301
list_x = list()
list_y = list()
con = get_connection()
if con.open():
    sql = 'select id_ticker from ticker where コード=%d;' % code
    query = QSqlQuery(sql)
    while query.next():
        id_ticker = query.value(0)
        print(id_ticker)
        break

    sql = 'select date, open from trade where id_code=%d;' % id_ticker
    query = QSqlQuery(sql)
    while query.next():
        list_x.append(query.value(0))
        list_y.append(query.value(1))

    con.close()
print(list_x)
print(list_y)
```
